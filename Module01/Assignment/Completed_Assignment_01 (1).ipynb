## Problem 1
(Problem adapted from scipy.com)

Stirling's Approximation for $ln(n!)$ is:

\begin{equation}
ln(n!) \approx nln(n) − n.
\end{equation}

Write a program to output $ln(n!)$ and its Stirling's approximation for $2 < n < 20$. Plot on the same graph the exact value and the approximate value as a function of n. Be sure to lable your curves and axes. **Hint**: the math module (and numpy.math) provides a factorial() method.

**Expected Plot (note this one has 2 < n < 100)**


```python
def ln_factorial(n):
    return math.log(math.factorial(n))

def stirling_approximation(n):
    return n * np.log(n) - n

n_vals = np.arange(3, 20)

ln_factorial_vals = []
stirling_vals = []

for n in n_vals:
    ln_factorial_vals.append(ln_factorial(n))
    stirling_vals.append(stirling_approximation(n))

plt.figure(figsize=(10, 6))
plt.plot(n_vals, ln_factorial_vals, label='Exact ln(n!)', color='blue', marker='o')
plt.plot(n_vals, stirling_vals, label='Stirling\'s Approximation', color='red', linestyle='--')

plt.xlabel('n')
plt.ylabel('Value')
plt.title('Comparison of ln(n!) and Stirling\'s Approximation')
plt.legend()
```

Now modify your program to output the exact value of $ln(n!)$, Stirling's approximation for it, and the relative error in the approximation for the first value of n where the releative error is less than 1%. The relative error in percent is defined as 100$\times$(exact-approx)/exact. Your output should quote the numbers to 3 decimal places. **Hint**: You may need to increase your n value and you may get an error message. Can you determine why you get an error and how to correctly make this comparison? 

```python
def relative_error(exact, approx):
    return 100 * abs((exact - approx) / exact)

# Initialize n starting from 3 and loop to find the first n where relative error < 1%
n = 3
threshold = 1  # threshold of 1% for the relative error
found = False

# Loop until the relative error is less than 1%
while not found:
    try:
        ln_fact = log(factorial(n))
        stirling_approx = n * log(n) - n
        error = relative_error(ln_fact, stirling_approx)
        
        # Check if the relative error is below the threshold
        if error < threshold:
            found = True
        else:
            n += 1  # Increment n and continue the search
            
    except OverflowError:
        print("Overflow error occurred at n = {}. Adjusting n or the approach may be necessary.".format(n))
        break

# Output the results for the first n where the relative error is less than 1%
if found:
    print("For n = {}, the exact value of ln(n!) is {:.3f}, Stirling's approximation is {:.3f}, "
          "and the relative error is {:.3f}%.".format(n, ln_fact, stirling_approx, error))the tested range.")
```
